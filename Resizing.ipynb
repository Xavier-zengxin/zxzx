```python
from imutils import *
```

```python
image = imread('image.jpg')
picture_show(image)
print(image.shape)
```

```python
width = 150
height = 150
image = cv2.resize(image,(width,height))
picture_show(image)
print(image.shape)
```

```python
image = imread('image.jpg')
width = 150
height = int(image.shape[0]*width/image.shape[1])
image = cv2.resize(image,(width,height))
picture_show(image)
print(image.shape)
```

```python
#最邻近算法
image = imread('image.jpg')
width = 150
height = int(image.shape[0]*width/image.shape[1])
image = cv2.resize(image,(width,height), interpolation = cv2.INTER_NEAREST)
picture_show(image)
print(image.shape)
```

```python
#双线性算法
image = imread('image.jpg')
width = 150
height = int(image.shape[0]*width/image.shape[1])
image = cv2.resize(image,(width,height), interpolation = cv2.INTER_LINEAR)
picture_show(image)
print(image.shape)
```

```python
#基于像素区域算法
image = imread('image.jpg')
width = 150
height = int(image.shape[0]*width/image.shape[1])
image = cv2.resize(image,(width,height), interpolation = cv2.INTER_AREA)
picture_show(image)
print(image.shape)
```

```python
#立方差值算法
image = imread('image.jpg')
width = 150
height = int(image.shape[0]*width/image.shape[1])
image = cv2.resize(image,(width,height), interpolation = cv2.INTER_CUBIC)
picture_show(image)
print(image.shape)
```

```python
#兰索斯差值算法
image = imread('image.jpg')
width = 150
height = int(image.shape[0]*width/image.shape[1])
image = cv2.resize(image,(width,height), interpolation = cv2.INTER_LANCZOS4)
picture_show(image)
print(image.shape)
```
